```python
pip install beautifulsoup4
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.canada.ca/en/health-canada/services/drugs-medication/cannabis/industry-licensees-applicants/licensed-cultivators-processors-sellers.html"
data = requests.get(url).content
soup = BeautifulSoup(data, 'html.parser')
'''find classes in the table so we can pick the right table'''
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
    
table = soup.find('table', class_='small')
table
```

```python
'''create the dataframe and define column headers'''
df = pd.DataFrame(columns=['License holder', 
                           'Province/territory',
                           'Licenses',
                           'Distributors/retailers',
                           'Patients',
                           'Client Care phone numbers',
                           'Date of initial licensing'])

lst = []


for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    
    if(columns != []):
        License_holder = columns[0].text.strip() 
        Province_territory = columns[1].text.strip() 
        Licenses = columns[2].text.strip().split(',')
        for l in Licenses:
            lst.append(l)
        Distributors_retailers = columns[3].text.strip().split('\n') 
        Patients = columns[4].text.strip().split('\n')
        Client_Care_phone_numbers = columns[5].text.strip() 
        Date_of_initial_licensing = columns[6].text.strip()
        
        df = df.append({'License holder': License_holder,  
                        'Province/territory': Province_territory,
                        'Licenses': Licenses,
                        'Distributors/retailers': Distributors_retailers, 
                        'Patients': Patients, 
                        'Client Care phone numbers': Client_Care_phone_numbers, 
                        'Date of initial licensing': Date_of_initial_licensing}, ignore_index=True)
set(lst)
```
